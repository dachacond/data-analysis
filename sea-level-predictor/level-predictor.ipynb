## Sea Level Predictor

You will analyze a dataset of the global average sea level change since 1880. You will use the data to predict the sea level change through year 2050.

Use the data to complete the following tasks:

- Use Pandas to import the data from epa-sea-level.csv.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv("epa-sea-level.csv")
df.head(10)
```

- Use matplotlib to create a scatter plot using the Year column as the x-axis and the CSIRO Adjusted Sea Level column as the y-axis.

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.scatter(df['Year'],df['CSIRO Adjusted Sea Level'], s=8)
ax.set_xlabel('year')
ax.set_ylabel('CSIRO Adjusted Sea Level')
```

- Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit. Plot the line of best fit over the top of the scatter plot. Make the line go through the year 2050 to predict the sea level rise in 2050.

```python
reg = linregress(df['Year'],df['CSIRO Adjusted Sea Level'])
```

```python
x = np.arange(df['Year'].min(), 2050)
ax.plot(x, reg.slope*x + reg.intercept, color='red')
fig
```

- Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset. Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as it has since the year 2000.

```python
df_sub = df[df['Year']>=2000]
```

```python
reg2 = linregress(df_sub['Year'],df_sub['CSIRO Adjusted Sea Level'])
x2 = np.arange(df_sub['Year'].min(), 2050)
ax.plot(x2, reg2.slope*x2 + reg2.intercept, color='blue')
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Rise in Sea Level')
fig
```
